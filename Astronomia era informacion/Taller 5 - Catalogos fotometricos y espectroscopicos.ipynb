```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.misc import imread, imsave
import IPython.display
%matplotlib inline
```

```python
data_url="https://raw.githubusercontent.com/rpmunoz/PentaUC/master/Astronomia%20era%20informacion/data/SDSS_DR12_phot_specz.csv"
data=pd.read_csv(data_url, sep=',', comment='#', header=0)
data.head()
```

```python
xrange=[15,23]
yrange=[-3,8]

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
plt.scatter(data['r'],data['u']-data['r'], color='blue', edgecolor='None', alpha=0.5)
plt.xlim(xrange)
plt.ylim(yrange)
```

```python
gv=(data['class']=='GALAXY')
print 'Minimum redsift: ', np.min(data['redshift'][gv])
print 'Maximum redshift: ', np.max(data['redshift'][gv])
```
